### Question 1 :

Pour deux signaux aléatoires $s_{1}(t)$ et $s_{2}(t)$, la covariance est:<br />
$cov\left ( s_{1}(t),s_{2}(t) \right ) = \mathbb{E}\left [ (s_{1}(t)-m_{1}(t)) (s_{2}(t)-m_{2}(t))\right ]$ <br />
$= \mathbb{E}\left [ s_{1}(t)s_{2}(t)\right ] + m_{1}(t)m_{2}(t) - m_{1}(t)\mathbb{E}\left [s_{2}(t)\right ] - m_{2}\mathbb{E}\left [s_{1}(t)\right ]$<br />
$ = \mathbb{E}\left [ s_{1}(t)s_{2}(t)\right ] - m_{1}(t)m_{2}(t)$<br />
$ = \mathbb{E}\left [ s_{1}(t)s_{2}(t)\right ] - \mathbb{E}\left [s_{1}(t)\right ]\mathbb{E}\left [s_{2}(t)\right ]$ <br /> <br />
si $\mathbb{E}\left [ s_{1}(t)s_{2}(t)\right ]=\mathbb{E}\left [s_{1}(t)\right ]\mathbb{E}\left [s_{2}(t)\right ]$ <br />
$s_{1}(t)$ et $s_{2}(t)$ sont non corrélés.<br />

Quant à la notion de l'indépendance, pour tout instant $t$ donné et deux signaux aléatoires $s_{1}(t)$ et $s_{2}(t)$, leur densité de probabilité jointe vérifie que <br />
$\rho(s_{1}(t)s_{2}(t)) = \rho(s_{1}(t))\rho(s_{2}(t))$ On dit que deux signaux $s_{1}$ et $s_{t}$  sont indépendants à l'instant $t$. <br /><br />

Dans le cas des signaux gaussiens, les notions de l'indépendance et de non corrélation sont égales.

### Question 9

Pour deux variables aléatoires indépendantes $s_{1}$ et $s2$, la densité de probabilité de la somme de ces deux variables aléatoires $s1+s2$ est égale au produit convolutionnel de leurs densité de probabilité $f_{s_{1}+s2} = f_{s_{1}}*f_{s2}$.<br />
$$
f_{s_{1}+s_{2}}(u) =
\begin{cases}
\frac{1}{12}\int_{-\sqrt{3}}^{\sqrt{3}+u} 1\times 1 dt = \mathbf{\frac{1}{12}(u+2\sqrt{3})}, \, -2\sqrt{3}\leqslant u \leqslant 0;\\
\frac{1}{12}\int_{-\sqrt{3}+u}^{\sqrt{3}} 1\times 1 dt = \mathbf{\frac{1}{12}(-u+2\sqrt{3})}, \, 0\leqslant u \leqslant 2\sqrt{3}; \\
0, ailleurs.
\end{cases}
$$<br />
Pour la variable alétoire $(s_{1}+s_{2})$, on peut obtenir son éspérance et sa variance: <br />
$\mathbb{E}[s_{1}+s_{2}] = \mathbb{E}[s_{1}] + \mathbb{E}[s_{2}] = 0$<br />
$var(s_{1}+s_{2}) = var(s_{1}) + var(s_{2}) = \frac{1}{12}(\sqrt(3)-(-\sqrt(3))^{2}\times 2 = 2$ ($s_{1}$ et $s_{2}$ sont indépendantes.)<br />

De la même manière, $f_{s_{1}+s_{2}+s_{3}} = f_{(s_{1}+s_{2})}*f_{s_{3}} $. On obtient : <br />
$$
f_{s_{1}+s_{2}+s_{3}}(u) =
\begin{cases}
\frac{1}{24\sqrt(3)}\int_{-2\sqrt{3}}^{\sqrt{3}+u} (t+2\sqrt{3}) \,dt = \mathbf{\frac{1}{24\sqrt(3)}[\frac{(u+\sqrt{3})^{2}}{2} + 2\sqrt{3}u +12]}, -3\sqrt{3}\leqslant u \leqslant -\sqrt{3};\\
\frac{1}{24\sqrt(3)}\left [ \int_{-\sqrt{3}+u}^{0}(t+2\sqrt{3})\,dt \,\,+\,\,\int_{0}^{\sqrt{3}+u}(2\sqrt{3}-t)\,dt \right ] = \mathbf{\frac{1}{24\sqrt(3)}(-u^{2}+9)}, -\sqrt{3}\leqslant u \leqslant \sqrt{3}; \\
\frac{1}{24\sqrt(3)}\int_{-\sqrt{3}+u}^{2\sqrt{3}} (-t+2\sqrt{3}) \,dt = \mathbf{\frac{1}{24\sqrt(3)}[\frac{(u-\sqrt{3})^{2}}{2} - 2\sqrt{3}u +12]}, \sqrt{3}\leqslant u \leqslant 3\sqrt{3};\\
0, ailleurs.
\end{cases}
$$<br />

Pour la variable alétoire $(s_{1}+s_{2}+s_{3})$, on peut obtenir son éspérance et sa variance: <br />
$\mathbb{E}[s_{1}+s_{2}+s_{3}] = \mathbb{E}[s_{1}] + \mathbb{E}[s_{2}] + \mathbb{E}[s_{3}] = 0$<br />

### Question 14 ICA POUR SÉPARER ET RETROUVER DEUX SIGNAUX

```python
% pylab inline
import numpy as np
import struct
#Q14
# from ICA.ipython
def source_separation_ICA(v_ica_in_1, v_ica_in_2,  d_fact_ech=1):
    
    nb             = len(v_ica_in_1)
    nb_iter        = 10**3
    m_melange      = array([v_ica_in_1 , v_ica_in_2])
    
    m_melange[0,:] = m_melange[0,:] - mean(m_melange[0,:])
    m_melange[1,:] = m_melange[1,:] - mean(m_melange[0,:])

    # whitening
    C    = dot(m_melange,m_melange.T)/nb
    d,V  = eig(C)
    M    = dot(dot(V, diag(1/sqrt(real(d)))), V.T)
    m_xw = dot(M, m_melange)
    
    # Search of one independant component
    v_w = rand(2)         # initialization of separation vector for first source
    v_w  = v_w/norm(v_w) 
    v_y1 = dot(v_w,m_xw)  # Initialization of estimated signals
    for i in range(nb_iter):           # estimation of signal entropy
        v_y    = tanh(v_y1)            #g(w^Tx)
        v_z    = 1/((cosh(v_y1))**2);  #g'(w^Tx)
        v_w[0] = mean(v_y*(m_xw[0,:]))-mean(v_z)*v_w[0];  #E[xg(w^Tx)]-E[g'(w^Tx)]w
        v_w[1] = mean(v_y*(m_xw[1,:]))-mean(v_z)*v_w[1];
        v_w    = v_w/norm(v_w);
        v_y1   = dot(v_w,m_xw)#v_w*m_xw.T;

    # search of second line seconde ligne.
    v_z = [-v_w[1], v_w[0]]
    m_b = [v_w, v_z]
        
    m_signal_estime = dot(m_b,m_xw) # estaimated signals

    output_1 = (m_signal_estime[0, :]) * d_fact_ech
    output_2 = (m_signal_estime[1, :]) * d_fact_ech
    
    return output_1,output_2

# Perform sources separation
output_1, output_2 = source_separation_ICA(x_1, x_2)

# Save outputs in files
file_output_1 = open("ICA_Out_1.txt","wb")
file_output_2 = open("ICA_Out_2.txt","wb")
file_output_1.write(struct.pack('d'*len(output_1),*output_1)) # write vector after binary conversion
file_output_2.write(struct.pack('d'*len(output_2),*output_2))
file_output_1.close()
file_output_2.close()

# Read reference signal files
reference_1 = np.fromfile("Ref_1.txt", dtype=numpy.float64)
reference_2 = np.fromfile("Ref_2.txt", dtype=numpy.float64)

# plot inputs and outputs
x_axis=linspace(0,220500,220500)
fig1 = plt.subplot(2, 2, 1)
fig1.set_title('recovered signal 1', fontsize=14, fontweight='bold')
plot(x_axis, output_1[0:220500])
xticks(np.arange(min(x_axis), max(x_axis)+1, 110000))

fig2 = plt.subplot(2, 2, 2)
fig2.set_title('reference signal 1', fontsize=14, fontweight='bold')
plot(x_axis,reference_1[0:220500])
xticks(np.arange(min(x_axis), max(x_axis)+1, 110000))
plt.tight_layout()

fig3 = plt.subplot(2, 2, 3)
fig3.set_title('recovered signal 2', fontsize=14, fontweight='bold')
plot(x_axis, output_2[0:220500])
xticks(np.arange(min(x_axis), max(x_axis)+1, 110000))

fig4 = plt.subplot(2, 2, 4)
fig4.set_title('reference signal 2', fontsize=14, fontweight='bold')
plot(x_axis,reference_2[0:220500])
xticks(np.arange(min(x_axis), max(x_axis)+1, 110000))
plt.tight_layout()

# Evaluate performance indices
cos_1   =  inner(reference_1,output_1)/norm(reference_1)/norm(output_1);
cos_2   =  inner(reference_2,output_2)/norm(reference_2)/norm(output_2);
err_1   = 10*(log(1-(cos_1**2))/log(10));
err_2   = 10*(log(1-(cos_2**2))/log(10));
print("Error index 1 (dB) = {:f}".format(err_1));
print("Error index 2 (dB) = {:f}\n".format(err_2));
print("cos(input_1, output_1) = {:f}".format(cos_1));
print("cos(input_2, output_2) = {:f}\n".format(cos_2));
```

```python
% pylab inline
from numpy import linalg as LA
import numpy as np
x_1  = np.fromfile("In_1.txt", dtype=np.float64)
x_2  = np.fromfile("In_2.txt", dtype=np.float64)
x_1 = np.asarray(x_1, dtype= np.float64)
x_2 = np.asarray(x_2, dtype= np.float64)
x = np.asarray([x_1,x_2], dtype= np.float64)
# zero mean
signaux_means = np.mean(x, axis=1)
normalisation = sqrt(2./(var(x_1)+var(x_2)))
x = x*normalisation
x = x - signaux_means[:,None]
nb = float(x.shape[1])
x_1 =reshape(x[0,:],(int(nb),1)) #vecteur colonne
x_2 =reshape(x[1,:],(int(nb),1)) #vecteur colonne
# correlation
#R_11 = dot(x_1, x_1.T); # shape = 220500**2 !! pas possible de  faire le produit matriciel
#R_12 = dot(x_1, x_2.T)/nb;
#R_22 = dot(x_2, x_2.T)/nb;
size_block = 10 # = "timedelay" in code C.
R1 = np.zeros([size_block,size_block], dtype=np.float64)
R2 = np.zeros([size_block,size_block], dtype=np.float64)
R12 = np.zeros([size_block,size_block], dtype=np.float64)
nb_blocks = int(nb/size_block)
for i in range(nb_blocks):
    begin = size_block*i
    end = size_block*i + size_block
    R1 += dot(x_1[begin:end],x_1[begin:end].T)/nb_blocks
    R2 += dot(x_2[begin:end], x_2[begin:end].T)/nb_blocks
    R12 += dot(x_1[begin:end],x_2[begin:end].T)/nb_blocks
    
T1 = np.trace(R1)/size_block
T2 = np.trace(R2)/size_block
T12 = np.trace(R12)/size_block
F1 = (np.sum(R1)-np.trace(R1))/(size_block-1)/size_block
F2 = (np.sum(R2)-np.trace(R2))/(size_block-1)/size_block
F12 = (np.sum(R12)-np.trace(R12))/(size_block-1)/size_block
Alpha = 2*F12*T12 - (F1*T2 + F2*T1)
Beta = 2*(T12**2 - T1*T2)
gamma2 = (F1*T2 - F2*T1)**2 + 4*(F12*T2-T12*F2)*(F12*T1 - T12*F1)
d1 = Alpha - sqrt(gamma2)
d2 = Alpha + sqrt(gamma2)

A_hat = np.array([[Beta*F1-T1*d1, Beta*F12-T12*d2], [Beta*F12 - T12*d1, Beta*F2-T2*d2]], dtype = np.float32)
A_inv = LA.inv(A_hat)
print ("A = ",A_hat)
```

```python
import struct
# suite du code précédent, A_inv est calculée avec le code précédent. 
# x = As => s = A_inv x
s = dot(A_inv, x)

# Read reference signal files
reference_1 = np.fromfile("Ref_1.txt", dtype=numpy.float64)
reference_2 = np.fromfile("Ref_2.txt", dtype=numpy.float64)

# plot inputs and outputs
x_axis=linspace(0,220500,220500)
fig1 = plt.subplot(2, 2, 1)
fig1.set_title('recovered signal 1', fontsize=14, fontweight='bold')
plot(x_axis, s[0,:])
xticks(np.arange(min(x_axis), max(x_axis)+1, 110000))

fig2 = plt.subplot(2, 2, 2)
fig2.set_title('reference signal 1', fontsize=14, fontweight='bold')
plot(x_axis,reference_1[0:220500])
xticks(np.arange(min(x_axis), max(x_axis)+1, 110000))
plt.tight_layout()

fig3 = plt.subplot(2, 2, 3)
fig3.set_title('recovered signal 2', fontsize=14, fontweight='bold')
plot(x_axis, s[1,:])
xticks(np.arange(min(x_axis), max(x_axis)+1, 110000))

fig4 = plt.subplot(2, 2, 4)
fig4.set_title('reference signal 2', fontsize=14, fontweight='bold')
plot(x_axis,reference_2[0:220500])
xticks(np.arange(min(x_axis), max(x_axis)+1, 110000))
plt.tight_layout()

# Evaluate performance indices
cos_1   =  inner(reference_1,s[0,:])/norm(reference_1)/norm(s[0,:]);
cos_2   =  inner(reference_2,s[1,:])/norm(reference_2)/norm(s[1,:]);

err_1   = 10*(log(1-(cos_1**2))/log(10));
err_2   = 10*(log(1-(cos_2**2))/log(10));
print ("timedelay = ", size_block)
print("Error index 1 (dB) = {:f}".format(err_1));
print("Error index 2 (dB) = {:f}\n".format(err_2));
print("cos(input_1, output_1) = {:f}".format(cos_1));
print("cos(input_2, output_2) = {:f}\n".format(cos_2));

output_1_sobi = s[0,:]/max(s[0,:])
output_2_sobi = s[1,:]/max(s[1,:])
# Save outputs in files
file_output_1 = open("SOBI_Out_1.txt","wb")
file_output_2 = open("SOBI_Out_2.txt","wb")
file_output_1.write(struct.pack('d'*len(output_1_sobi),*output_1_sobi)) # write vector after binary conversion
file_output_2.write(struct.pack('d'*len(output_2_sobi),*output_2_sobi))
file_output_1.close()
file_output_2.close()
```
